```python
import feedparser

# 解析 RSS 源
# 人工智能

rss_url = 'https://www.google.com/alerts/feeds/17855623974115894960/3515207334378775940'  # 替换为你的 RSS 源 URL
feed = feedparser.parse(rss_url)

# 打印 RSS 源的标题和更新时间
print(f"Feed Title: {feed.feed.title}")
print(f"Updated: {feed.feed.updated}")

# 遍历并打印每个条目的标题、链接、描述和发布日期
for entry in feed.entries:
    print(f"ID: {entry.id}")
    print(f"Title: {entry.title}")
    print(f"Link: {entry.link}")
    print(f"Published: {entry.published}")
    print(f"Content: {entry.content[0].value}")
    print("-" * 40)
```
